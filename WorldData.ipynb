Configuring TabPy Server:

Since I am are on a Tableau free trial, I might not have a Tableau Server. In that case, you can use TabPy in "External Service" mode. Hence, I will download and install TabPy server on your local machine. You can find it on the official GitHub repository. (https://github.com/tableau/TabPy/blob/master/docs/server-install.md)

To start TabPy server on your machine, these were the commands on my machine:
* python -m pip install --upgrade pip
* pip install tabpy
* tabpy

*Note: If you want to add config you can look at the GitHub link to find out.*

Connecting Tableau Desktop with TabPy:

Open Tableau Desktop.

Go to Help > Settings and Performance > Manage External Service Connection.

In the "External Service Connection" dialog, enter the following information:
* Server: localhost (or the IP address of your machine)
* Port: 9004 (or the port on which TabPy is running)
* Authentication: None (for local setup)

Click Test Connection to ensure it's successful.
